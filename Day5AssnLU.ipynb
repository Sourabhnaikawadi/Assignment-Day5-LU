 Q1 sort given list and keep all zeros at right side

```
list = [0,1,2,10,4,1,0,56,2,0,1,3,0,56,0,4]
list.sort()
print(list)
```

```
 
# moving element to end  
 
list.append(list.pop(list.index(0))) 
 
  
# printing result 
 
print ("The modified element moved list is : " + str(list))
```

Q2 merge and sort two lists using only one loop

```
list1 = [10,20,40,60,70,80]
list2 = [5,15,25,35,45,60]
res = sorted(list1 + list2) 
 
  
# printing result 
 
print ("The combined sorted list is : " + str(res))
```
